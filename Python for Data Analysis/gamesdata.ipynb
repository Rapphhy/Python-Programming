```python
#import dataset and view header
import numpy as np
import pandas as pd
df = pd.read_csv('vgsales.csv')
df.head()
```

```python
#checking data overview 
df.info()
```

```python
# Statistical overview
df.describe()
```

```python
#checking for missing values
df.isnull().sum()
```

```python
#Removing duplicates
df.dropna(inplace=True)
print(df.info())
```

```python
# Top Platform
# First we sort by Platform and sum the Global_Sales values, then we arrange this in descending order.
top_platform = df.groupby('Platform', as_index=False)['Global_Sales'].sum().sort_values(by='Global_Sales', ascending=False) 
print(top_platform)
```

```python
#Exploratory Data Analysis
#Total Sales over the years
import matplotlib.pyplot as plt
df.groupby("Year")["Global_Sales"].sum().plot(kind="line", figsize=(12,5), title = 'Total Sales over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales')
plt.show()

#Note: we can see a consistent trend below 100, between 1980 to 1985. followed by an up trend to 2010, when the chart starts to decline.
```

```python
# Sales Trend by region
import matplotlib.pyplot as plt

df.groupby("Year")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].sum().plot(kind="line", figsize=(12,5), title="Global Sales Over the Years")
plt.xlabel("Year")
plt.ylabel("Global Sales")
plt.legend(["North America", "Europe", "Japan", "Other"])
plt.show()
# From the chart, every country maintain sales below 50 between 1980 - 1995. Then there was a spike in North america which contributed mostly to the rise in 2010.
```

```python
# Most played game
df["Genre"].value_counts().plot(kind="bar", figsize=(12,5), title="Game Distribution")
plt.xlabel("Genre")
plt.ylabel("Number of Games")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group data by Genre, summing sales for each region
genre_sales = df.groupby("Genre")[["NA_Sales", "EU_Sales", "JP_Sales"]].sum()

# Compute total sales and select the top 10 genres
genre_sales["Global_Sales"] = genre_sales.sum(axis=1)
top_10_genres = genre_sales.nlargest(10, "Global_Sales").reset_index()

# Define size for the 3 charts 
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Define a function to plot bar charts and assign region 
def plot_top_genres(ax, sales_column, region, color):
    sns.barplot(y=top_10_genres["Genre"], x=top_10_genres[sales_column], ax=ax)
    ax.set_title(f"Top 10 Best-Selling Games in {region}")
    ax.set_xlabel(f"{region} Sales")
    ax.set_ylabel("Genre")

# graph plot
plot_top_genres(axes[0], "NA_Sales", "North America", "Blues")
plot_top_genres(axes[1], "EU_Sales", "Europe", "Greens")
axes[2].pie(top_10_genres["JP_Sales"], labels=top_10_genres["Genre"], autopct="%1.1f%%", colors=sns.color_palette("Reds", len(top_10_genres)))
axes[2].set_title("Top 10 Best-Selling Games in Japan")

plt.tight_layout()
plt.show()
```

```python
# Top Publishers per region
# import matplotlib.pyplot as plt
import numpy as np
publisher_sales = df.groupby("Publisher")[["NA_Sales", "EU_Sales", "JP_Sales"]].sum()

# Compute total sales and select the top 10 publishers
publisher_sales["Global_Sales"] = publisher_sales.sum(axis=1)
top_10_publishers = publisher_sales.nlargest(10, "Global_Sales").reset_index()

# Set positions for the x-axis
x = np.arange(len(top_10_publishers)) 
width = 0.25  # Width of each bar

# graph plot
fig, ax = plt.subplots(figsize=(12, 5))

ax.bar(x - width, top_10_publishers["NA_Sales"], width, label="North America", color="blue")
ax.bar(x, top_10_publishers["EU_Sales"], width, label="Europe")
ax.bar(x + width, top_10_publishers["JP_Sales"], width, label="Japan", color="grey")

# Formating the chart
ax.set_xlabel("Publisher", fontsize=12)
ax.set_ylabel("Global Sales", fontsize=12)
ax.set_title("Top 10 Publishers", fontsize=14, fontweight="bold")
ax.set_xticks(x)
ax.set_xticklabels(top_10_publishers["Publisher"], rotation=45, ha="right", fontsize=10)
ax.legend(title="Region")

plt.show()
```
